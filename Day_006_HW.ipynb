```python
## 作業
# 將下列部分資料片段 sub_train 使用 One Hot encoding, 並觀察轉換前後的欄位數量 (使用 shape) 與欄位名稱 (使用 head) 變化；

# ---->轉換前後欄位數量由307511 rows × 1 columns 變成 307511 rows × 7 columns
# 欄位名稱由 WEEKDAY_APPR_PROCESS_START變成 WEEKDAY_APPR_PROCESS_START_FRIDAY, WEEKDAY_APPR_PROCESS_START_MONDAY,
# WEEKDAY_APPR_PROCESS_START_SATURDAY, WEEKDAY_APPR_PROCESS_START_SUNDAY,	WEEKDAY_APPR_PROCESS_START_THURSDAY,
# WEEKDAY_APPR_PROCESS_START_TUESDAY 以及 WEEKDAY_APPR_PROCESS_START_WEDNESDAY
```

```python
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
print(sub_train.shape)
sub_train.head()
```

```python
data_dum = pd.get_dummies(sub_train)
pd.DataFrame(data_dum)
```

```python
# 另一種
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
sub_train['WEEKDAY_APPR_PROCESS_START'] = labelencoder.fit_transform(sub_train['WEEKDAY_APPR_PROCESS_START'] )

sub_train
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(categories='auto')
data_str_ohe = onehotencoder.fit_transform(sub_train).toarray()
pd.DataFrame(data_str_ohe)
```

```python
# 另一種
# 使用one-hot encoding 讀	WEEKDAY_APPR_PROCESS_START這一欄的資料轉成one hot ecoding
from tqdm.auto import tqdm
cls = {
       "SUNDAY":0, "MONDAY":1, "TUESDAY":2, "WEDNESDAY":3,
       "THURSDAY":4, "FRIDAY":5, "SATURDAY":6
       }
label = np.empty((307511,),dtype=np.float32)
for i, column in tqdm(enumerate(app_train['WEEKDAY_APPR_PROCESS_START'])):
  label[i] = cls[column]
```

```python
print(label[0])
print(app_train["WEEKDAY_APPR_PROCESS_START"][0])
```

```python
# one-hot-encoding只可以做一次喔
from tensorflow.keras.utils import to_categorical
one_hot_encoding = to_categorical(label,num_classes=7)
```
